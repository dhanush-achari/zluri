# List the production house for every year from 2000-2016 which has released the most number of movies in that year

```python
df1 = new_df.copy()
df1['release_date'].head()
```

#### Lets convert string value to date and parse only the year

```python
df1['release_date'] = pd.to_datetime(df1['release_date'])
df1['release_year'] = pd.DatetimeIndex(df1['release_date']).year
df1['release_year'].head()
```

#### To get the count of all the production lets binarize each production house using label binarizer

```python
from sklearn.preprocessing import LabelBinarizer
labelbinarizer = LabelBinarizer()
output = labelbinarizer.fit_transform(df1['production_house'])
df_bin = pd.DataFrame(output,columns=labelbinarizer.classes_)
df_bin.drop('',axis=1,inplace=True)
df_bin.tail()
```

#### Merge with df1 and perform groupby on date

```python
df1 = pd.merge(df1,df_bin,how="outer",left_index=True,right_index=True)
df1 = df1.loc[:,'release_year':]
df1.sort_values("release_year",inplace=True)
```

#### Only considering values between 2000 - 2016
#### Also has a null value to be handled

```python
df1.isnull().any(axis=1)
# THERE IS ONE null Value in release year so lets drop it
df1 = df1.dropna(how='any')
# Always put brackets while comparing with '&' and '|'
df1 = df1[(df1['release_year']>=2000) & (df1['release_year'] <=2016)]
df1
```

```python
df1 = df1.groupby("release_year").sum()
df1.sample(5)
```

```python
# df1.apply(lambda x:x.argmax ,axis = 1)
df1.apply(lambda x: x.argmax(), axis=1)
```
