# 2. top 5 production houses (by budget) and list their top 5 most popular movies, their revenue and vote_average


### Lets convert string back to list 

```python
base_df['production_companies'][0]
```

```python
from ast import literal_eval
```

```python
base_df['production_companies'] = base_df['production_companies'].apply(lambda x : literal_eval(x) )
```

```python
base_df['production_companies'][0]
```

```python
base_df.columns
```

#### Lets create a DataFrame with only necessary columns so that it is cleaner looking

```python
df_nes = base_df[['budget',
       'original_title','popularity', 'production_companies'
       , 'release_date', 'revenue', 'runtime',
        'status', 'title', 'vote_average',
       'vote_count']]
df_nes.head()
```

#### Lets parse only the production companies and remove other junk


```python
df_nes['production_companies'] = df_nes['production_companies'].apply(lambda x : [i['name'] for i in x] )
df_nes
```

```python
lst = df_nes.sort_values('budget',ascending=False).head(5)['production_companies'].to_list()
lst
```

### Top 5 production companies based on budget


```python
for i in lst:
    st=''
    for j in i:
        st+=','+j
    print(st[1:])
```

```python
new_df = df_nes
new_df.head()
```

## Since Top 5 production houses is asked in the question 
#### Let us parse all the production houses

```python
df=new_df.production_companies.apply(pd.Series)
```

```python
df = df.fillna('')
df = df.dropna(how="all")
df
```

```python
df['production_house'] =  df.apply(lambda x:','.join([str(i) for i in x if i != '']), axis=1)
```

```python
df
```

```python
new_df['production_house'] = df['production_house']
new_df.head()
```

```python
new_df['production_house'].value_counts()
```

## Top 5 individual production house according budget
#### Lets groupby using prodution house

```python
new_df.groupby('production_house').agg({'budget':np.sum}).sort_values('budget',ascending=False).iloc[1:6,:]
```
