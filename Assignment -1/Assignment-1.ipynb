# 3. List the production house for every year from 2000-2016 which has released the most number of movies in that year

```python
df1 = new_df.copy()
df1['release_date'].head()
```

#### Lets convert string value to date and parse only the year

```python
df1['release_date'] = pd.to_datetime(df1['release_date'])
df1['release_year'] = pd.DatetimeIndex(df1['release_date']).year
df1['release_year'].head()
```

#### To get the count of all the production lets binarize each production house using label binarizer

```python
from sklearn.preprocessing import LabelBinarizer
labelbinarizer = LabelBinarizer()
output = labelbinarizer.fit_transform(df1['production_house'])
df_bin = pd.DataFrame(output,columns=labelbinarizer.classes_)

df_bin.drop('',axis=1,inplace=True) #DROP THE COLUMN HAVING NO PRODUTION HOUSE NAME
df_bin.tail()
```

#### Merge with df1 and perform groupby on date

```python
df1 = pd.merge(df1,df_bin,how="outer",left_index=True,right_index=True)
df1 = df1.loc[:,'release_year':]
df1.sort_values("release_year",inplace=True)
df1.head(5)
```

#### Only considering values between 2000 - 2016
#### Also has a null value to be handled

```python
df1.isnull().any(axis=1)
# THERE IS ONE null Value in release year so lets drop it
# Interpolate could have been used but it may skew the data as we are trying to find the prodution house hence we drop
df1 = df1.dropna(how='any')
# Always put brackets while comparing with '&' and '|'
df1 = df1[(df1['release_year']>=2000) & (df1['release_year'] <=2016)]
df1
```

```python
df1 = df1.groupby("release_year").sum()
df1
```

# 4. You are going to invest all your life's savings in a production company. You have two choices "Marvel Studios" and "DC Comics". Which company would you bet on? This is an open-ended question. Define your own metrics to measure which one is a better investment opportunity and defend your analysis.

```python
df_invest = new_df.copy()
df_invest.head(5)
```

#### Creating column for profits of each prodution company
#### Creating columns for both and marking 1 when they produced a movie else 0

```python
df_invest['Profit'] = df_invest['revenue'] - df_invest['budget']
df_invest['Marvel']=df_invest['production_companies'].apply(lambda x : 1 if 'Marvel Studios' in x else 0)
df_invest['DC']=df_invest['production_companies'].apply(lambda x : 1 if 'DC Comics' in x else 0)
```

```python
df_invest.sample(5)
```

#### Creating df for movies produced by marvel and dc and calculating their profits 

```python
df_marvel = df_invest[df_invest['Marvel']==1]
df_dc = df_invest[df_invest['DC']==1]
```

#### The average popularity profit generated by Marvel and DC per year

```python
df_marvel['release_date'] = pd.to_datetime(df_marvel['release_date'])
df_marvel['release_year'] = pd.DatetimeIndex(df_marvel['release_date']).year
df_marvelyear = df_marvel.groupby("release_year").agg({'Profit':np.average,'popularity':np.average})
df_marvelyear
```

```python
df_dc['release_date'] = pd.to_datetime(df_dc['release_date'])
df_dc['release_year'] = pd.DatetimeIndex(df_dc['release_date']).year
df_dcyear = df_dc.groupby("release_year").agg({'Profit':np.average,'popularity':np.average})
df_dcyear
```

#### To understand their profit pefomance better lets make some visualizations

```python
import matplotlib.pyplot as plt
%matplotlib inline
%matplotlib notebook
plt.style.use(['seaborn'])
```

```python
years = list(range(1966,2017,4))
plt.figure(figsize=(10,5))
plt.plot(df_marvelyear['Profit'],'r-',label="Marvel profit/year",marker='.')
plt.plot(df_dcyear['Profit'],'b-',label="DC profit/year",marker='.')
plt.xticks(years)
plt.xlabel('Years',fontsize=14)
plt.ylabel('Profit in Billions',fontsize=14)
plt.title("Profit VS Years",fontsize=20)

plt.legend()
```

# Conclusion 1:
### From the above line graph it clearly evident that at present the average profit generated by MARVEL per year is way larger compared to DC. Thus i would hands down invest in MARVEL , as the prodution house that generates more profits would yeild better RETURNS to me

#### If i was born in the 1960s i might have invested in DC since there was LINEAR growth in profit generated by DC from 1966 - 2005 and also i there was no option for MARVEL

```python
years = list(range(1966,2017,4))
plt.figure(figsize=(10,5))
plt.plot(df_marvelyear['popularity'],'r-',label="Marvel popularity/year")
plt.plot(df_dcyear['popularity'],'b-',label="DC popularity/year",marker='.')
plt.xticks(years)
plt.xlabel('Years',fontsize=14)
plt.ylabel('Populatrity',fontsize=14)
plt.title("Populatrity VS Years",fontsize=20)

plt.legend()
```
